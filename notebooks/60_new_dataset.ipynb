## Prepare Data

```python
import json

j = 0
contains_rumex = []
root_centers = []
bounding_boxes = []
image_names = []

wrong_labeld = []

with open('../data/labels.json') as json_file:  
    data = json.load(json_file)
    for i, subdata in enumerate(data):
        if i == 0:
            continue
        try:
            if subdata['Label']['bild_enthaelt_ampfer'] == 'ja':
                subdata['Label']['Wurzelpunkt']
        except:
            wrong_labeld.append(subdata['External ID'])
            continue
        if subdata['Label']['bild_enthaelt_ampfer'] == 'ja':
            r_c = []
            b_b = []
            for p in subdata['Label']['Wurzelpunkt']:
                r_c.append(p)
            for p in subdata['Label']['Ampfer']:
                b_b.append(p)
            contains_rumex.append(subdata['Label']['bild_enthaelt_ampfer'])
            root_centers.append(r_c)
            bounding_boxes.append(b_b)
            image_names.append(subdata['External ID'])
        if subdata['Label']['bild_enthaelt_ampfer'] == 'nein':
            r_c = []
            b_b = []
            contains_rumex.append(subdata['Label']['bild_enthaelt_ampfer'])
            root_centers.append(r_c)
            bounding_boxes.append(b_b)
            image_names.append(subdata['External ID'])

assert len(contains_rumex) == len(root_centers) == len(bounding_boxes) == len(image_names)

print("Wrong Labeled Images:")
for l in wrong_labeld:
    print(l)
```

```python

columns = {"contains_rumex":[ True if c=='ja' else False for c in contains_rumex ],
         "root_centers":root_centers,
          "bounding_boxes" : bounding_boxes,
          "image_names" : image_names}
test_field_df = pd.DataFrame.from_dict(columns)
test_field_df
```
